```python
#import stuff
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
#read csv
df = pd.read_csv("iris.csv")
df
```

```python
from sklearn.cluster import KMeans
predict = KMeans(n_clusters = 3, random_state = 0, n_init = "auto").fit(df[['sepallength', 'sepalwidth', 'petallength', 'petalwidth']])
```

```python
predict.get_feature_names_out()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(figsize=(12, 12))
axs.scatter(df['petallength'], df['petalwidth'], c=predict.labels_)
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(12, 12))
axs.scatter(df['sepallength'], df['sepalwidth'], c=predict.labels_)
plt.show()
```

```python
from sklearn.metrics import jaccard_score
from sklearn.metrics import rand_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import silhouette_score
from sklearn.metrics import davies_bouldin_score




# Assuming df is your feature matrix
labels = predict.labels_

rand = rand_score(df['class'], labels)
print(f"Rand Score: {rand}")


calins = calinski_harabasz_score(df[['sepallength', 'sepalwidth', 'petallength', 'petalwidth']], labels)
print(f"Calinski & Harabasz Score: {calins}")


silhouette = silhouette_score(df[['sepallength', 'sepalwidth', 'petallength', 'petalwidth']], labels)
print(f"Silhouette Score: {silhouette}")


davies = davies_bouldin_score(df[['sepallength', 'sepalwidth', 'petallength', 'petalwidth']], labels)
print(f"Davies Bouldin Index: {davies}")

```
